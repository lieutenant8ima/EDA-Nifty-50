```python
#Grouping On Company Names Column & Open Column And Applying mathematical mean().Hence we can get company wise avg value for 
#Open share points
om=data.groupby(['Company Names'])['Open'].mean()

print("Mean Of Open Col In Data Table Group By Company Names Are", om )
```

```python
#Above Calculated Mean Of companies Showed Using BoxPlot,IQR is Positively Skewed That. 
plt.figure()
open_mean= sns.boxplot(x=om ,data=data, orient="v")
```

```python
#Grouping On Company Names Column & Close Column And Applying mathematical mean().Hence we can get company wise avg value for 
#Close share points

cm=data.groupby(['Company Names'])['Close'].mean()

print("Mean Of Close Col In Data Table Group By Company Names Are", cm )
```

```python
#Above Calculated Mean Of companies Showed Using BoxPlot,IQR is Positively Skewed That.
closed_mean = sns.boxplot(x=cm,data=data, orient="v")
```

```python
#Finding Mean Of Highest Share Point By Applying GroupBy Fn On Company Names
h=data.groupby(['Company Names'])['High'].mean()

print("Mean Of High's In Data Table Group By Company Names Are", h )
```

```python
#Above Calculated Mean Of companies Showed Using BoxPlot,IQR is Positively Skewed That.
highest= sns.boxplot(x=h,data=data, orient="v")
```

```python
#Finding Mean Of Lowest Share Point By Applying GroupBy Fn On Company Names
l=data.groupby(['Company Names'])['Low'].mean()

print("Mean Of Low's In Data Table Group By Company Names Are", l )
```

```python
#Above Calculated Mean Of companies Showed Using BoxPlot,IQR is Positively Skewed That.
lowest = sns.boxplot(x=l,data=data, orient="v")
```
